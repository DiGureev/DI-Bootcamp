```python
starbucks['caffeine_to_calories'] = starbucks['Caffeine (mg)'].replace('Varies','0').replace('varies','0').astype(float)/starbucks['Calories']

starbucks
```

```python
starbucks.value_counts('Beverage_category','caffeine_to_calories')
```

```python
starbucks.value_counts('Beverage_category','caffeine_to_calories').plot.barh(x='Beverage_category', y='caffeine_to_calories')
```
